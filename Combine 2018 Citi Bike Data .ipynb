```python
# file = os.path.join("Data/201801-citibike-tripdata.csv")
jan18 = pd.read_csv("Data/citibike 0118.csv")
feb18 = pd.read_csv("Data/citibike 0218.csv")
mar18 = pd.read_csv("Data/citibike 0318.csv")
apr18 = pd.read_csv("Data/citibike 0418.csv")
may18 = pd.read_csv("Data/citibike 0518.csv")
jun18 = pd.read_csv("Data/citibike 0618.csv")
jul18 = pd.read_csv("Data/citibike 0718.csv")
aug18 = pd.read_csv("Data/citibike 0818.csv")
sep18 = pd.read_csv("Data/citibike 0918.csv")
oct18 = pd.read_csv("Data/citibike 1018.csv")
nov18 = pd.read_csv("Data/citibike 1118.csv")
dec18 = pd.read_csv("Data/citibike 1218.csv")
```

```python
months = jan18
allMonths = [feb18, mar18, apr18, may18, jun18, jul18, aug18, sep18, oct18, nov18, dec18]
for data in allMonths:
    months = pd.concat([months, data], axis = 0)
months.head()
```
